```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the data
data = pd.read_csv('../data/MachineLearningRating_v3.txt', sep='|', low_memory=False)

# Display the first few rows of the DataFrame
print(data.head())

# Check the data types of each column
print(data.dtypes)


```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing values:")
print(missing_values)
```

```python
# Handle missing values
# For numerical columns, fill missing values with mean or median
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
data[numerical_cols] = data[numerical_cols].fillna(data[numerical_cols].median())
print(data[numerical_cols])
```

```python
# For categorical columns, fill missing values with mode
categorical_cols = data.select_dtypes(include=['object']).columns
data[categorical_cols] = data[categorical_cols].fillna(data[categorical_cols].mode().iloc[0])
print(data[categorical_cols])
```

```python
# Univariate Analysis
# Distribution of Numerical Variables
plt.figure(figsize=(10, 6))
sns.histplot(data['TotalPremium'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of TotalPremium')
plt.xlabel('TotalPremium')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data['TotalClaims'], bins=30, kde=True, color='salmon')
plt.title('Distribution of TotalClaims')
plt.xlabel('TotalClaims')
plt.ylabel('Frequency')
plt.show()

# Distribution of Categorical Variables
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Gender', palette='Set2')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='MaritalStatus', palette='Set1')
plt.title('Distribution of MaritalStatus')
plt.xlabel('MaritalStatus')
plt.ylabel('Count')
plt.show()
```

```python
# Task 1.3: Univariate Analysis

# Distribution of numerical variables
plt.figure(figsize=(12, 6))
sns.histplot(data['TotalPremium'], bins=20, kde=True, color='blue')
plt.title('Distribution of TotalPremium')
plt.xlabel('TotalPremium')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(data['TotalClaims'], bins=20, kde=True, color='green')
plt.title('Distribution of TotalClaims')
plt.xlabel('TotalClaims')
plt.ylabel('Frequency')
plt.show()

# Bar charts for categorical variables
plt.figure(figsize=(12, 6))
sns.countplot(data['CoverCategory'], palette='Set1')
plt.title('Count of Cover Categories')
plt.xlabel('Cover Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Task 1.4: Bivariate or Multivariate Analysis
# Correlation Matrix
correlation_matrix = data[['TotalPremium', 'TotalClaims']].corr()
print(correlation_matrix)

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='TotalPremium', y='TotalClaims', hue='Province', palette='Set1')
plt.title('Scatter Plot of TotalPremium vs. TotalClaims by Province')
plt.xlabel('TotalPremium')
plt.ylabel('TotalClaims')
plt.legend(title='PostalCode')
plt.show()

# Correlation matrix
correlation_matrix_zip = data.groupby('Province')[['TotalPremium', 'TotalClaims']].corr().unstack()['TotalClaims']['TotalPremium']

print("Correlation Matrix:")
print(correlation_matrix_zip)
```

```python
#Another  correlation Analysis
# Select relevant columns
relevant_columns = ['TotalPremium', 'TotalClaims', 'MaritalStatus', 'Gender', 'Province']
relevant_data = data[relevant_columns]

# Convert categorical variables into dummy/indicator variables (one-hot encoding)
relevant_data = pd.get_dummies(relevant_data, columns=['MaritalStatus', 'Gender', 'Province'])

# Compute the correlation matrix
correlation_matrix = relevant_data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Matrix of Relevant Parameters')
plt.show()
```

```python
# Task 1.5: Data Comparison
# Plot 1: Bar Plot of Cover Type by Province
plt.figure(figsize=(12, 6))
sns.countplot(x='Province', hue='CoverType', data=data, palette='Set1')
plt.title('Cover Type Distribution by Province')
plt.xlabel('Province')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Cover Type')
plt.show()

# Plot 2: Box Plot of Total Premium by Province
plt.figure(figsize=(12, 6))
sns.boxplot(x='Province', y='TotalPremium', data=data)
plt.title('Total Premium Distribution by Province')
plt.xlabel('Province')
plt.ylabel('Total Premium')
plt.xticks(rotation=45)
plt.show()

# Plot 3: Bar Plot of Top Auto Makes by Province
top_auto_makes = data['make'].value_counts().head(5).index
top_auto_data = data[data['make'].isin(top_auto_makes)]

plt.figure(figsize=(12, 6))
sns.countplot(x='Province', hue='make', data=top_auto_data, palette='Set2')
plt.title('Top Auto Makes by Province')
plt.xlabel('Province')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Auto Make')
plt.show()
```
